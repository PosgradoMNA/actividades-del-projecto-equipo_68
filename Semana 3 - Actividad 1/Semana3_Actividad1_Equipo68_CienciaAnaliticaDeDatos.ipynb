# *Parte 1: Fundamentos de bases de datos*



1. Fundamentos de bases de datos y para ciencia de datos. 

Una base de datos es una recopilación de datos lógicamente relacionados y organizados en forma de una tabla enorme compuesta por columnas, filas y celdas. 
Las columnas representan las categorías de la base de datos, y la primera fila está asignada a los nombres de estas categorías, mientras que las filas son cada una de las entradas, por ejemplo, cada cliente, cada empleado, etc, lo que aumenta la eficacia del proceso y la consulta de los datos. 
Una base de datos permite al usuario observar, imprimir y mostrar los datos a conveniencia con el objetivo de informar al usuario u organización.  


2. Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.

Un almacén de datos es un repositorio centralizado que almacena datos estructurados, y semiestructurados para que estos puedan ser analizados. Un almacén de datos es algo así como una gran base de datos que contiene información proveniente de varios orígenes y desde muchos años atrás hasta datos más actuales. Un almacén de datos protege, limpia y transforma los datos para que sean privados, precisos y seguros y para que de esta forma los datos puedan ser utilizados para tomar decisiones más inteligentes e informadas, además de que hace que el tiempo de consulta sea más corto.

**Solución 2**

Imputación de media:

```python
nTCClientes = TCClientes.copy()

#Ciclo para obtener la media en cada columna y sustituir los valores NaN por la media.
for label in  nTCClientes.columns:          #En este ciclo se mueve sobre todas las columnas de data frame, desde X1 a Y   
  nTCClientes[label].fillna(nTCClientes[label].mean(), inplace = True)    #En cada iteración se obtiene la media de la columna iterada por medio del 
                                                                          #método .mean() y este es imputado(asignado) a los valores NaN por medio de la
                                                                          #función .fillna()




print(nTCClientes.isnull().any())         #Confirmación de que no hay NaNs
print(nTCClientes.mean())                 #Mostrar las medias de cada columna
nTCClientes
```

Imputación de mediana:

```python
nTCClientes = TCClientes.copy()

#Ciclo para obtener la mediana en cada columna y sustituir los valores NaN por la mediana.
for label in  nTCClientes.columns:          #En este ciclo se mueve sobre todas las columnas de data frame, desde X1 a Y   
  nTCClientes[label].fillna(nTCClientes[label].median(), inplace = True)    #En cada iteración se obtiene la mediana de la columna iterada por medio del 
                                                                          #método .mean() y este es imputado(asignado) a los valores NaN por medio de la
                                                                          #función .fillna()




print(nTCClientes.isnull().any())         #Confirmación de que no hay NaNs
print(nTCClientes.median())                 #Mostrar las medias de cada columna
nTCClientes
```

Se logra ya no tener NaNs pero en las columnas X6 a X11, hay valores de 0, los cuales no significan nada o que no está dentro de la escala mencionada en la descripción. Se procede a cambiar a -1 asumiendo que estos clientes no tienen ningun adeudo.

```python
nTCClientes.X6= nTCClientes.X6.replace([0,-1])
nTCClientes.X7= nTCClientes.X7.replace([0,-1])
nTCClientes.X8= nTCClientes.X8.replace([0,-1])
nTCClientes.X9= nTCClientes.X9.replace([0,-1])
nTCClientes.X10= nTCClientes.X10.replace([0,-1])
nTCClientes.X11= nTCClientes.X11.replace([0,-1])
nTCClientes
```

**Generar un nuevo dataframe sin NaN y con formato nuevo:**

En el siguiente codigo se generará un nuevo data frame, sin NaNs, estos seran remplazados conforme a la siguiente tabla para evitar valores fuera de escala. Ademas se renombraran los valores de los encabezados para mejor manejo de los datos.

image.png

Al analizar el data frame original, se detectan valores extranios en algunas columnas, por ejemplo X6, X7, ..., X11, valores como -2 y 0, estos valores se cambian a -1, asumiendo que estas personas no son deudores. Para la educación y el estado marital se remplaza por la moda.
 Ya que 0, es un valor fuera de la escala como se muestra en el documento de la descripción.

image.png

```python
#Creación del data frame:
nTCClientes = TCClientes.copy()

nTCClientes['X3'].replace(to_replace=[0,5,6], value = nTCClientes['X3'].mode()[0], inplace = True)
nTCClientes['X4'].replace(0,nTCClientes['X4'].mode()[0], inplace = True)
nTCClientes['X6'].replace(to_replace=[-2,0], value=-1, inplace = True)
nTCClientes['X7'].replace(to_replace=[-2,0], value=-1, inplace = True)
nTCClientes['X8'].replace(to_replace=[-2,0], value=-1, inplace = True)
nTCClientes['X9'].replace(to_replace=[-2,0], value=-1, inplace = True)
nTCClientes['X10'].replace(to_replace=[-2,0], value=-1, inplace = True)
nTCClientes['X11'].replace(to_replace=[-2,0], value=-1, inplace = True)
```

```python
#Ciclo para remplazar los NaNs, para ciertas columnas se imputa la Moda, para mantener valores dentro de la escala,
#en los demas se imputa la media, conforme la tabla anterior.

for label in  nTCClientes.columns:    
  if label == 'X2' or label == 'X3' or label == 'X4' or label == 'X6' or label == 'X7' or label == 'X8' or label == 'X9' or label == 'X10' or label == 'X11' or label == 'Y':
      #Imputación de moda: 
      nTCClientes[label].fillna(nTCClientes[label].mode()[0], inplace = True)   
  else:
      #Imputación de media:
      nTCClientes[label].fillna(nTCClientes[label].mean(), inplace = True)
```

```python
#Renombrar los encabezados de las columnas:
nTCClientes.rename(columns = {'X1' :'Credit_Amount',      
                              'X2' : 'Gender', 
                              'X3': 'Education', 
                              'X4':'Marital_Status',
                              'X5': 'Age',
                              'X6': 'Sep',
                              'X7': 'Ago',
                              'X8': 'Jul',
                              'X9': 'Jun',
                              'X10': 'May',
                              'X11': 'Apr',
                              'X12': 'Bs-Sep',
                              'X13': 'Bs-Ago',
                              'X14': 'Bs-Jul',
                              'X15': 'Bs-Jun',
                              'X16': 'Bs-May',
                              'X17': 'Bs-Apr',
                              'X18': 'Pp-Sep',
                              'X19': 'Pp-Ago',
                              'X20': 'Pp-Jul',
                              'X21': 'Pp-Jun',
                              'X22': 'Pp-May',
                              'X23': 'Pp-Apr',
                              'Y':'Credible?'
                              }, inplace = True) 

nTCClientes.to_csv('data.csv')            #Se guarda este nuevo data frame como data.csv en la raiz de Colab 
```

incorrectos a datos que no lo sean realmente, ocasionando que se eliminen más
datos que los que deberían.
